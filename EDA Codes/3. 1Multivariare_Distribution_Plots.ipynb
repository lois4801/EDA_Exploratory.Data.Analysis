```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```
tips = sns.load_dataset('tips')

tips.head()
```

```
fig, ax = plt.subplots(figsize=(10.5, 7.2))

sns.scatterplot(x='total_bill', y='tip', data=tips, color='#F15B2A', ax=ax)
```

```
sns.jointplot(x='total_bill', y='tip', data=tips, color='#F15B2A', kind='kde')
```

```
fig = plt.figure(figsize=(21, 14.4))
sns.kdeplot(tips['total_bill'], tips['tip'])
```

```
from mpl_toolkits.mplot3d import Axes3D
```

```
fig = plt.figure(figsize=(10.5, 7.2))
ax = fig.gca(projection='3d')

ax.plot_trisurf(tips['total_bill'], tips['tip'], tips['size'], color='#F15B2A')
```

```
fig = plt.figure(figsize=(15, 9))
ax = fig.add_subplot(111, projection='3d')

ax.bar3d(tips['tip'], tips['total_bill'], np.zeros(len(tips['size'])), 1, 1, tips['size'], shade=True, color='#F15B2A') 
```

```
fig, ax = plt.subplots(figsize=(10.5, 7.2))

sns.boxplot(x='time', y='tip', hue='smoker', data=tips, ax=ax)
```

```
fig, ax = plt.subplots(figsize=(10.5, 7.2)) 

sns.violinplot(x='time', y='tip', hue='smoker', data=tips, ax=ax)
```
